#### Project 5: Online Food Ordering Analysis

This notebook explores customer demographics and behavior from an online food ordering platform.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df_food = pd.read_csv('onlinefoods.csv')
print("Online foods dataset loaded successfully!")
```

#### 1. Customer Age Distribution
A histogram to understand the age range of the customer base.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_food['Age'], kde=True, color='purple', bins=20)
plt.title('Age Distribution of Online Food Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#### 2. Monthly Income vs. Repeat Orders
This visualization explores whether a customer's monthly income level affects their likelihood of placing a repeat order.

```python
plt.figure(figsize=(12, 7))
income_order = ['No Income', '10001 to 25000', '25001 to 50000', 'More than 50000']
sns.countplot(y='Monthly Income', hue='Output', data=df_food, order=income_order, palette='magma')
plt.title('Relationship between Monthly Income and Repeat Orders')
plt.xlabel('Number of Customers')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.show()
```

```python
# ---Customer Feedback ---
plt.figure(figsize=(8, 8))
feedback_counts = df_food['Feedback'].value_counts()
plt.pie(feedback_counts, labels=feedback_counts.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 14})
plt.title('Distribution of Customer Feedback', fontsize=16)
plt.ylabel('')
plt.show()
```

Key Insights: Online Food Orders

1. Target Audience: Young Adults and Students
The age distribution histogram shows a massive spike in customers between the ages of **22 and 25**. This indicates with high certainty that the primary market for this online food service is composed of college students and young professionals who are just starting their careers.

2. Customer Satisfaction is Extremely High
The pie chart reveals that **over 75% of all feedback is positive**. This is a very strong signal of a healthy service with high customer satisfaction. The low incidence of negative feedback suggests the business is consistently meeting or exceeding customer expectations regarding food quality, delivery, and overall experience.
