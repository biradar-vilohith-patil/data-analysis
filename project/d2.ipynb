# Project 2: Analysis of College Placement Data
- analyzes a dataset of college placements to understand factors affecting whether a student gets placed.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df_placement = pd.read_csv('./CollegePlacement.csv')
print("Placement dataset loaded successfully!")
```

##### 1. Placement Status Overview
overall distribution of placed vs. not placed students.

```python
plt.figure()
sns.countplot(x='Placement', data=df_placement)
plt.title('Overall Placement Status')
plt.xlabel('Placed or Not (1 = Placed, 0 = Not Placed)')
plt.ylabel('Number of Students')
plt.show()
```

##### 2. Placement by Academic Stream
placement status by the student's academic result.

```python
plt.figure()
sns.violinplot(x='Prev_Sem_Result', hue='Placement', data=df_placement)
plt.title('Placement Status by previous result')
plt.xlabel('GPA')
plt.ylabel('Number of Students')
plt.legend(title='Status', labels=['No', 'Yes'])
plt.show()
```

```python
# --- Internship Experience vs Placement ---
plt.figure(figsize=(10, 6))
sns.countplot(data=df_placement, x='Internship_Experience', hue='Placement')
plt.title('Impact of Internship Experience on Placement', fontsize=16)
plt.xlabel('Has Internship Experience?')
plt.ylabel('Number of Students')
plt.legend(title='Placement Status')
plt.show()
```

- Key Insights: College Placement Analysis

1. CGPA is a Critical Factor for Placement
The box plot clearly demonstrates that students who successfully secured a placement have a **significantly higher median CGPA**. The interquartile range for placed students is entirely above the median for unplaced students, indicating that strong and consistent academic performance is a primary requirement for employers.

2. Internship Experience Provides a Decisive Edge
While it's possible to get placed without an internship, the data shows that having internship experience provides a **strong competitive advantage**. The ratio of placed to unplaced students is dramatically more favorable for those who have completed an internship, highlighting the value employers place on practical, real-world experience.
