#### DataSet - most_runs_average_strikerate.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```python
#Load dataset
df = pd.read_csv("./most_runs_average_strikerate.csv")
```

```python
# Get information about the columns and data types
df.info()
```

```python
# Get descriptive statistics for the numerical columns
df.describe()
```

##### Visualizations

```python
# Find the top 10 run scorers
top_10_runs = df.nlargest(10, 'total_runs')

plt.figure(figsize=(12, 7))
sns.barplot(x='total_runs', y='batsman', data=top_10_runs)
plt.title('Top 10 Run Scorers in IPL')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.show()
```

```python
# Analyze the top 50 players for better visualization
top_50_players = df.nlargest(50, 'total_runs')

plt.figure(figsize=(14, 9))
sns.scatterplot(x='average', y='strikerate', size='total_runs', hue='batsman', data=top_50_players, sizes=(50, 600), legend=False)
plt.title('Average vs. Strike Rate for Top 50 IPL Batsmen')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.axvline(top_50_players['average'].mean(), color='r', linestyle='--', label='Mean Average')
plt.axhline(top_50_players['strikerate'].mean(), color='b', linestyle='--', label='Mean Strike Rate')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
min_2000_runs = df[df['total_runs'] > 2000]
top_10_strikerate = min_2000_runs.nlargest(10, 'strikerate')
sns.barplot(data=top_10_strikerate, y='strikerate', x='batsman')
plt.title('Top 10 Strike Rates (min. 2000 Runs)', fontsize=16)
plt.xlabel('Strike Rate')
plt.ylabel('Batsman')
plt.show()
```

Key Insights: Cricket Batsmen Performance

1. The "Accumulators": Consistency and Longevity
The chart for total runs is dominated by iconic, long-serving players like **V Kohli and SK Raina**. Their position at the top reflects their remarkable consistency and ability to score a high volume of runs season after season.

2. The "Power Hitters": Scoring Speed
The strike rate chart tells a different story. When we filter for players who have scored a significant number of runs (to ensure a meaningful sample size), players like **CH Gayle and DA Warner** rise to the top. This effectively separates the fast scorers from the accumulators and highlights the players teams rely on for quick, impactful innings.
