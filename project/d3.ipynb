#### 1. Feature Correlation
A heatmap is a great way to see how different health metrics are correlated with each other and with the target variable (presence of heart disease).

```python
sns.heatmap(df_heart.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Heart Disease Features')
plt.show()
```

```python
# ---  Histogram of Median Home Value (MEDV) ---
plt.figure(figsize=(10, 6))
sns.histplot(df_heart['MEDV'], kde=True, bins=30)
plt.title('Distribution of Median Home Value (MEDV)', fontsize=16)
plt.xlabel('Median Value in $1000s')
plt.ylabel('Number of Houses')
plt.show()
```

```python

# --- Visualization 2: Scatter Plot of Rooms vs. Price ---
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_heart, x='RM', y='MEDV', alpha=0.7)
plt.title('Average Number of Rooms vs. Median Home Value', fontsize=16)
plt.xlabel('Average Number of Rooms (RM)')
plt.ylabel('Median Value in $1000s (MEDV)')
plt.show()
```

```python
# ---Correlation Heatmap ---
plt.figure(figsize=(14, 10))
correlation_matrix = df_heart.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Housing Features', fontsize=16)
plt.show()
```

Key Insights: Boston Housing Dataset Analysis

1. House Prices Are Normally Distributed, With a Catch
The histogram of median home values (`MEDV`) shows a relatively normal distribution, with most house prices clustering around the **$20,000 to $25,000** range. 
2. More Rooms = Higher Price (A Strong Linear Relationship)
The scatter plot provides a very clear and intuitive insight: there is a **strong, positive, and linear relationship** between the average number of rooms (`RM`) and a house's median value (`MEDV`). 
3. Neighborhood Status is a Dominant Factor
While the number of rooms is important, the correlation heatmap reveals another, more powerful factor. The `LSTAT` variable—which represents the percentage of the population that is considered "lower status"—has the **strongest negative correlation (-0.74)** with the median home value.
