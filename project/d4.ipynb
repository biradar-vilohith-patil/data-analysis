#### Project 1: Analysis of the Coffee & Health Dataset

This notebook explores the synthetic dataset on coffee consumption and its relationship with health metrics like cholesterol and blood pressure.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent style for all plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load the dataset
df_coffee = pd.read_csv('synthetic_coffee_health_10000.csv')
print("Coffee dataset loaded successfully!")
```

1. Initial Data Exploration
Let's start by inspecting the data to understand its structure, data types, and summary statistics.

```python
# Display the first few rows of the dataframe
df_coffee.head()
```

```python
# Get information about the columns and data types
df_coffee.info()
```

```python
# Get descriptive statistics for the numerical columns
df_coffee.describe()
```

#### 2. Visualizations
Now, let's create some visualizations to uncover patterns.

##### Distribution of Coffee Cups Per Day
A histogram to see how many cups of coffee people drink daily.

```python
plt.figure()
sns.histplot(df_coffee['Coffee_Intake'], kde=True, bins=10)
plt.title('Distribution of Coffee Cups Per Day')
plt.xlabel('Coffee Cups Per Day')
plt.ylabel('Frequency')
plt.show()
```

##### Coffee Consumption vs. Sleep
A scatter plot to see if there's a relationship between coffee intake and sleep hours.

```python
# --- Coffee Intake vs. Sleep Hours ---
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_coffee, x='Coffee_Intake', y='Sleep_Hours', alpha=0.4)
plt.title('Coffee Intake (Cups/Day) vs. Sleep Hours', fontsize=16)
plt.xlabel('Daily Coffee Intake (Cups)')
plt.ylabel('Hours of Sleep')
plt.show()
```

```python
# --- Sleep Quality vs. Stress Level ---
plt.figure(figsize=(10, 6))
sns.countplot(data=df_coffee, x='Stress_Level', hue='Sleep_Quality', order=['Low', 'Medium', 'High'])
plt.title('Sleep Quality Breakdown by Stress Level', fontsize=16)
plt.xlabel('Stress Level')
plt.ylabel('Count of Individuals')
plt.show()
```

Key Insights: Coffee and Health Data

#### 1. Coffee's Impact on Sleep is Weaker Than Expected
The scatter plot shows a very weak negative correlation between coffee intake and sleep hours. While there's a slight downward trend, the points are widely dispersed. This suggests that while very high consumption can affect sleep, for most people, **other factors are far more influential on sleep duration than the number of cups of coffee consumed**.

#### 2. Stress is the Real Enemy of Good Sleep
The relationship between stress and sleep quality is stark and undeniable. The bar chart shows that **"Good" sleep quality is almost exclusively found in the "Low" stress group**, while "Fair" sleep quality overwhelmingly dominates the "High" stress group. This powerful visual indicates that managing stress is the most critical factor for achieving a good night's sleep, far more so than modulating coffee intake.
